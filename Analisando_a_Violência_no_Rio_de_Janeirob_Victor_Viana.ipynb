###Segurança no setor automotivo 

* **Vamos Analisar a segurança no setor automototivo verificando algumas médias:**
  * Roubo de veículos 
  * Furto de Veículos 
  * Recuperação de Veículos 

```
# média do roubo de veículos 
print(" A média de veículos roubados é: {0:.2f} por mês".format(df.roubo_veiculo.mean()))
print(" A média de veículos furtados é: {0:.2f} por mês".format(df.furto_veiculos.mean()))
print(" A média de recuperação dos veículos é: {0:.2f} por mês".format(df.recuperacao_veiculos.mean()))
```

* **Além disso, é possível achar a taxa de recuperação de veículos**
 


```
taxa_recupera =df.recuperacao_veiculos.mean()/ (df.roubo_veiculo.mean() + df.furto_veiculos.mean() )
print("%.2f"% taxa_recupera)
```

A taxa de recuperação dos veículos é de 43%

## Conclusão

Esse é um Data Set muito extenso que possui várias informações sobre a violência no Rio de Janeiro, temos quase 30 anos de dados. Conseguimos tirar algumas informações importantes e que **impactam diretamente a vida dos indivíduos e do mercado.** 

O primeiro deles sendo o **aumento do roubo a transeuntes e carga** com o passar dos anos. O efeito negativo nos indivíduos é duplo, visto que com mais roubos de cargas seguros ficam mais caros, além da precificação considerar as perdas futuras. Assim, consumidores se deparam com **preços mais altos** devido a falta de segurança.

A **taxa de recuperação dos veículos** que foi encontrada é um dado interessante,porém teríamos que ter os dados de quantos desses 43% são recuperados e em consições de uso. Isso porque muitas vezes quando são recuperados, foram retiradas peças importantes para venda em mercado negro e o carro sofreu danos consideráveis que tornam seu uso algo impraticável.

A **média de roubo e furto de celular** a partir de 2010 também é muito relevante pois reflete os altos preços praticados pelo mercado para seguro de celulares. 

No entanto, é evidente que por ser um DataSet muito grande em termos de tempo e as cidades estão em constante mudança. Tal fato ficou evidente quando em vários gráficos de crescimento (de diferentes variáveis) as mudanças conteceram em períodos perecidos 2004/2005 e 2010/2011. Nesse sentido, para compreensão de dados históricos esse Data Set se mostra interesante porém se queremos compreender o momento atual e fazer previsões recomenda-se um Data Set com um intervalo mínimo de 10 anos.
